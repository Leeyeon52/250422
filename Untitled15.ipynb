```python
import pandas as pd
import numpy as np
import cv2
import os
from tqdm import tqdm

# ë°ì´í„° íŒŒì¼ ê²½ë¡œ ì„¤ì •
train_file = 'train.csv'
test_file = 'test.csv'
labels_file = 'sample_submission.csv'

# ì´ë¯¸ì§€ ì €ì¥ ê²½ë¡œ ì„¤ì •
image_dir = 'url_images'
annotation_dir = 'url_annotations'
test_image_dir = 'test_url_images'
os.makedirs(image_dir, exist_ok=True)
os.makedirs(annotation_dir, exist_ok=True)
os.makedirs(test_image_dir, exist_ok=True)

# ì´ë¯¸ì§€ í¬ê¸° ì„¤ì •
img_width = 256
img_height = 32

# ë¼ë²¨ ë§¤í•‘ (ì•…ì„±: 1, ì •ìƒ: 0)
label_map = {'benign': 0, 'malicious': 1}

# URLì„ ì´ë¯¸ì§€ë¡œ ë³€í™˜í•˜ëŠ” í•¨ìˆ˜
def url_to_image(url, width, height):
    """URL ë¬¸ìì—´ì„ ê°„ë‹¨í•œ ì´ë¯¸ì§€ë¡œ ë³€í™˜í•©ë‹ˆë‹¤."""
    img = np.zeros((height, width, 1), dtype=np.uint8)
    for i, char in enumerate(url):
        if i < width:
            img[height // 2, i, 0] = ord(char) % 256  # ASCII ê°’ì„ ì´ìš©í•˜ì—¬ í”½ì…€ ê°’ í• ë‹¹
    return img

def generate_yolo_annotation(label, img_width, img_height):
    """YOLO í˜•ì‹ì˜ annotation ìƒì„± (ë‹¨ì¼ ê°ì²´ ê°€ì •)."""
    # ê°ì²´ê°€ ì´ë¯¸ì§€ ì¤‘ì•™ì— ìˆë‹¤ê³  ê°€ì • (ë‹¨ìˆœí™”)
    x_center = 0.5
    y_center = 0.5
    width = 1.0
    height = 1.0
    return f"{label} {x_center} {y_center} {width} {height}"

try:
    # í•™ìŠµ ë°ì´í„° ë¡œë“œ ë° ì „ì²˜ë¦¬
    train_df = pd.read_csv(train_file)
    if 'label' in train_df.columns:
        train_df['label'] = train_df['label'].map(label_map)
        # 'label' ì»¬ëŸ¼ì— NaN ê°’ì´ ìˆëŠ” í–‰ ì œê±°
        train_df.dropna(subset=['label'], inplace=True)
        # ë¼ë²¨ì„ ì •ìˆ˜í˜•ìœ¼ë¡œ ë³€í™˜
        train_df['label'] = train_df['label'].astype(int)
    else:
        print("ì£¼ì˜: 'label' ì»¬ëŸ¼ì´ train.csvì— ì—†ìŠµë‹ˆë‹¤. ë¼ë²¨ ì •ë³´ í™•ì¸ í•„ìš”.")

    print("í•™ìŠµ ë°ì´í„° ì´ë¯¸ì§€ ë° annotation ìƒì„±...")
    for index, row in tqdm(train_df.iterrows(), total=len(train_df)):
        # URL ì»¬ëŸ¼ ì°¾ê¸°
        url_column = None
        for col in ['url', 'URL', 'address']:
            if col in row:
                url_column = col
                break
        if url_column is None:
            print(f"ì˜¤ë¥˜: URL ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. (ì¸ë±ìŠ¤: {index})")
            continue
        url = row[url_column]
        label = int(row['label'])
        image = url_to_image(url, img_width, img_height)
        image_filename = f"train_{index}.png"
        cv2.imwrite(os.path.join(image_dir, image_filename), image)

        annotation = generate_yolo_annotation(label, img_width, img_height)
        annotation_filename = f"train_{index}.txt"
        with open(os.path.join(annotation_dir, annotation_filename), 'w') as f:
            f.write(annotation)

    # í•™ìŠµ ë°ì´í„° ì´ë¯¸ì§€ ë° annotation ëª©ë¡ íŒŒì¼ ìƒì„± (YOLOv í•™ìŠµì— í•„ìš”)
    with open('train.txt', 'w') as f:
        for index in range(len(train_df)):
            image_filename = f"url_images/train_{index}.png"
            f.write(f"{os.path.abspath(image_filename)}\n")

    # í…ŒìŠ¤íŠ¸ ë°ì´í„° ë¡œë“œ ë° ì´ë¯¸ì§€ ìƒì„±
    test_df = pd.read_csv(test_file)
    print("í…ŒìŠ¤íŠ¸ ë°ì´í„° ì´ë¯¸ì§€ ìƒì„±...")
    for index, row in tqdm(test_df.iterrows(), total=len(test_df)):
        # URL ì»¬ëŸ¼ ì°¾ê¸°
        url_column = None
        for col in ['url', 'URL', 'address']:
            if col in row:
                url_column = col
                break
        if url_column is None:
            print(f"ì˜¤ë¥˜: í…ŒìŠ¤íŠ¸ ë°ì´í„°ì—ì„œ URL ì»¬ëŸ¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤. (ì¸ë±ìŠ¤: {index})")
            continue
        url = row[url_column]
        image = url_to_image(url, img_width, img_height)
        image_filename = f"test_{index}.png"
        cv2.imwrite(os.path.join(test_image_dir, image_filename), image)

    # í…ŒìŠ¤íŠ¸ ë°ì´í„° ì´ë¯¸ì§€ ëª©ë¡ íŒŒì¼ ìƒì„±
    with open('test.txt', 'w') as f:
        for index in range(len(test_df)):
            image_filename = f"test_url_images/test_{index}.png"
            f.write(f"{os.path.abspath(image_filename)}\n")

except FileNotFoundError as e:
    print(f"íŒŒì¼ì„ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {e}")
except KeyError as e:
    print(f"KeyError ë°œìƒ: {e}. CSV íŒŒì¼ì— í•´ë‹¹ ì»¬ëŸ¼ì´ ìˆëŠ”ì§€ í™•ì¸í•´ì£¼ì„¸ìš”.")
except ValueError as e:
    print(f"ValueError ë°œìƒ: {e}")

print("ë°ì´í„° ì¤€ë¹„ ì™„ë£Œ. YOLOv ëª¨ë¸ í•™ìŠµì„ ì§„í–‰í•  ìˆ˜ ìˆìŠµë‹ˆë‹¤.")
print("train.txt, test.txt, url_images, url_annotations í´ë”ë¥¼ í™•ì¸í•˜ì„¸ìš”.")
```
