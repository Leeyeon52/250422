```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

# 1. 데이터 불러오기 및 전처리
df = pd.read_csv("jena_climate_2009_2016.csv")
df = df[5::6]  # 10분 간격 데이터를 1시간 간격으로 다운샘플링

# 사용할 특성 선택
cols = ["T (degC)", "p (mbar)", "rho (g/m**3)"]
features = df[cols].copy()

# 날짜 파싱 (dayfirst 옵션 사용)
features.index = pd.to_datetime(df["Date Time"], dayfirst=True)

# 2. 정규화
train_mean = features[:200000].mean()
train_std = features[:200000].std()
features = (features - train_mean) / train_std

# 3. 시퀀스 데이터 생성 함수
def make_dataset(data, window_size, horizon):
    X = []
    Y = []
    for i in range(len(data) - window_size - horizon):
        x = data.iloc[i:i+window_size].values  # iloc을 사용하여 데이터 슬라이싱
        y = data.iloc[i + window_size + horizon - 1, 0]  # T(degC) 예측
        X.append(x)
        Y.append(y)
    return np.array(X), np.array(Y)

# 파라미터 설정
window_size = 24  # 과거 24시간
horizon = 1       # 1시간 뒤 예측

# 시퀀스 데이터 생성
X, Y = make_dataset(features, window_size, horizon)

# 데이터 분할
split1 = int(len(X) * 0.7)
split2 = int(len(X) * 0.9)
X_train, Y_train = X[:split1], Y[:split1]
X_val, Y_val = X[split1:split2], Y[split1:split2]
X_test, Y_test = X[split2:], Y[split2:]

# Tensorflow Dataset으로 변환
batch_size = 32
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, Y_train)).shuffle(1000).batch(batch_size)
val_dataset = tf.data.Dataset.from_tensor_slices((X_val, Y_val)).batch(batch_size)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, Y_test)).batch(batch_size)

# 4. 모델 정의
model = keras.models.Sequential([
    layers.LSTM(32, return_sequences=True, input_shape=(window_size, len(cols))),
    layers.LSTM(16),
    layers.Dense(1)
])

model.compile(loss="mse", optimizer="adam")
model.summary()

# 5. 모델 훈련
callbacks = [
    keras.callbacks.ModelCheckpoint("jena_best_model.keras", save_best_only=True)
]

history = model.fit(
    train_dataset,
    epochs=20,
    validation_data=val_dataset,
    callbacks=callbacks
)

# 6. 최적 모델 불러와서 평가
model = keras.models.load_model("jena_best_model.keras")
test_loss = model.evaluate(test_dataset)
print("Test Loss (MSE):", test_loss)

# 7. 예측 예시
for x_batch, y_batch in test_dataset.take(1):
    preds = model.predict(x_batch)
    print("실제값:", y_batch.numpy()[:5])
    print("예측값:", preds[:5].flatten())
```

```python
import os
import numpy as np
import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.optimizers import RMSprop, Adam
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# 업로드된 CSV 파일 이름
data_path = "jena_climate_2009_2016.csv"

# 데이터 로드 및 전처리
if not os.path.exists(data_path):
    print(f"오류: {data_path} 파일을 찾을 수 없습니다. 코랩에 파일이 업로드되었는지 확인해주세요.")
else:
    try:
        with open(data_path) as f:
            data = f.read()

        lines = data.split("\n")
        header = lines[0].split(",")
        lines = lines[1:]

        float_data = np.zeros((len(lines), len(header) - 1))
        for i, line in enumerate(lines):
            values = [float(x) for x in line.split(",")[1:]]
            float_data[i, :] = values

        # 데이터 정규화
        scaler = MinMaxScaler()
        scaled_data = scaler.fit_transform(float_data)

        # 시계열 데이터 생성 함수 (이전과 동일)
        def generator(data, lookback, delay, min_index, max_index, shuffle=False, batch_size=128, step=6):
            if max_index is None:
                max_index = len(data) - delay - 1
            i = min_index + lookback
            while True:
                if shuffle:
                    rows = np.random.randint(min_index + lookback, max_index, size=batch_size)
                else:
                    if i + batch_size >= max_index:
                        i = min_index + lookback
                    rows = np.arange(i, min(i + batch_size, max_index))
                    i += len(rows)

                samples = np.zeros((len(rows), lookback // step, data.shape[-1]))
                targets = np.zeros((len(rows),))
                for j, row in enumerate(rows):
                    indices = range(rows[j] - lookback, rows[j], step)
                    samples[j] = data[indices]
                    targets[j] = data[rows[j] + delay][1]  # 기온 예측

                yield samples, targets

        # 파라미터 설정 (이전과 동일)
        lookback = 720  # 120시간 전 (6시간 간격)
        step = 6
        delay = 144  # 24시간 후
        batch_size = 128

        # 훈련, 검증, 테스트 데이터 제너레이터 생성 (이전과 동일)
        train_gen = generator(scaled_data, lookback=lookback, delay=delay, min_index=0, max_index=200000, shuffle=True, step=step, batch_size=batch_size)
        val_gen = generator(scaled_data, lookback=lookback, delay=delay, min_index=200001, max_index=300000, step=step, batch_size=batch_size)
        test_gen = generator(scaled_data, lookback=lookback, delay=delay, min_index=300001, max_index=None, step=step, batch_size=batch_size)

        val_steps = (300000 - 200001) // batch_size
        test_steps = (len(scaled_data) - 300001 - lookback) // batch_size

        # 기준 모델 (단순 평균) (이전과 동일)
        def evaluate_naive_method():
            batch_maes = []
            for step in range(val_steps):
                samples, targets = next(val_gen)
                naive_preds = samples[:, -1, 1]
                mae = np.mean(np.abs(naive_preds - targets))
                batch_maes.append(mae)
            print(f"기준 모델 MAE: {np.mean(batch_maes):.2f}")

        evaluate_naive_method()

        # 학습 곡선 시각화 함수 (이전과 동일)
        def plot_loss(history, title):
            loss = history.history['loss']
            val_loss = history.history['val_loss']
            epochs = range(1, len(loss) + 1)
            plt.figure()
            plt.plot(epochs, loss, 'bo', label='Training loss')
            plt.plot(epochs, val_loss, 'b', label='Validation loss')
            plt.title(title)
            plt.xlabel('Epochs')
            plt.ylabel('Loss')
            plt.legend()
            plt.show()

        # 첫 번째 LSTM 모델 구축 및 학습 (이전과 동일)
        model1 = Sequential([
            layers.LSTM(32, input_shape=(None, scaled_data.shape[-1])),
            layers.Dense(1)
        ])

        optimizer = RMSprop()
        model1.compile(optimizer=optimizer, loss='mae')
        history1 = model1.fit(train_gen, steps_per_epoch=500, epochs=20, validation_data=val_gen, validation_steps=val_steps)

        plot_loss(history1, 'Initial LSTM Model Training and Validation Loss')

        # 나머지 모델 구축 및 학습 코드는 이전과 동일하게 이어집니다.
        # (model2, model3, model4, model5, model6, model7, model8)

    except Exception as e:
        print(f"데이터 로드 중 오류 발생: {e}")
```

```python
import glob
import numpy as np
import pandas as pd
from tensorflow.keras.utils import to_categorical
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Embedding, Reshape, Input
import tarfile
import os

# 압축 파일 경로
tgz_file = 'jsb_chorales.tgz'
extract_dir = '.'

try:
    if not os.path.exists('jsb_chorales'):
        with tarfile.open(tgz_file, 'r:gz') as tar:
            tar.extractall(extract_dir)
        print(f"'{tgz_file}' 파일이 성공적으로 압축 해제되었습니다. '{extract_dir}/jsb_chorales' 폴더를 확인하세요.")
    else:
        print(f"'{extract_dir}/jsb_chorales' 폴더가 이미 존재합니다. 압축 해제를 건너뜁니다.")
except FileNotFoundError:
    print(f"오류: '{tgz_file}' 파일을 찾을 수 없습니다. 파일이 현재 작업 디렉토리에 있는지 확인하세요.")
except Exception as e:
    print(f"압축 해제 중 오류 발생: {e}")

train_folder = 'jsb_chorales/train'
valid_folder = 'jsb_chorales/valid'
test_folder = 'jsb_chorales/test'

def load_data_from_folder(folder_path):
    sequences = []
    file_paths = glob.glob(f'{folder_path}/*.csv')
    for file_path in file_paths:
        try:
            df = pd.read_csv(file_path, header=None)
            sequences.append(df.values.tolist())
        except Exception as e:
            print(f"Error reading file {file_path}: {e}")
    return sequences

train_sequences = load_data_from_folder(train_folder)
valid_sequences = load_data_from_folder(valid_folder)
test_sequences = load_data_from_folder(test_folder)

all_notes = set()
for seq in train_sequences + valid_sequences + test_sequences:
    for step in seq:
        for note in step:
            all_notes.add(note)
all_notes = sorted(list(all_notes))
note_to_index = {note: i for i, note in enumerate(all_notes)}
index_to_note = {i: note for note, i in note_to_index.items()}
num_notes = len(all_notes)

def prepare_sequences(sequences, note_to_index, seq_length):
    input_sequences = []
    output_notes = []
    for seq in sequences:
        if len(seq) > seq_length:
            for i in range(len(seq) - seq_length):
                input_seq = seq[i:i + seq_length]
                output_note = seq[i + seq_length]
                input_sequences.append([note_to_index.get(note, -1) for step in input_seq for note in step])
                output_notes.append([note_to_index.get(note, -1) for note in output_note])
    input_sequences = [list(filter(lambda x: x != -1, seq)) for seq in input_sequences]
    output_notes = [list(filter(lambda x: x != -1, notes)) for notes in output_notes]

    valid_indices = [i for i, (inp, out) in enumerate(zip(input_sequences, output_notes)) if len(inp) == seq_length * 4 and len(out) == 4]
    input_sequences = np.array([input_sequences[i] for i in valid_indices])
    output_notes = np.array([output_notes[i] for i in valid_indices])

    return input_sequences, output_notes

seq_length = 32
X_train_raw, y_train_raw = prepare_sequences(train_sequences, note_to_index, seq_length)
X_valid_raw, y_valid_raw = prepare_sequences(valid_sequences, note_to_index, seq_length)

num_notes = len(note_to_index)

X_train = X_train_raw.reshape(-1, seq_length, 4) if X_train_raw.size > 0 else np.array([])
X_valid = X_valid_raw.reshape(-1, seq_length, 4) if X_valid_raw.size > 0 else np.array([])

y_train = np.array([to_categorical(y, num_classes=num_notes) for y in y_train_raw]) if y_train_raw.size > 0 else np.array([])
y_valid = np.array([to_categorical(y, num_classes=num_notes) for y in y_valid_raw]) if y_valid_raw.size > 0 else np.array([])

print("훈련 데이터 형태:", X_train.shape, y_train.shape)
print("검증 데이터 형태:", X_valid.shape, y_valid.shape)
print("총 음표 개수:", num_notes)

embedding_dim = 128
lstm_units = 256
num_features = 4 * embedding_dim

model = Sequential([
    Input(shape=(seq_length, 4)),
    Embedding(input_dim=num_notes, output_dim=embedding_dim),
    Reshape((seq_length, num_features)),
    LSTM(units=lstm_units, return_sequences=False),
    Dense(4 * num_notes, activation='softmax'),
    Reshape((4, num_notes))
])

model.compile(optimizer='adam', loss='categorical_crossentropy')
model.summary()

early_stopping = EarlyStopping(monitor='val_loss', patience=10, verbose=1, restore_best_weights=True)
model_checkpoint = ModelCheckpoint('best_model.keras', monitor='val_loss', save_best_only=True, verbose=1)

if X_train.size > 0 and y_train.size > 0:
    history = model.fit(
        X_train, y_train,
        epochs=100,
        batch_size=64,
        validation_data=(X_valid, y_valid),
        callbacks=[early_stopping, model_checkpoint]
    )
else:
    print("훈련 데이터가 없습니다. 데이터 로딩 및 전처리 과정을 확인하세요.")

def generate_music(model, seed_sequence, num_steps, note_to_index, index_to_note, num_notes):
    generated_sequence = [list(step) for step in seed_sequence]
    input_sequence = np.array([[note_to_index.get(note, 0) for note in step] for step in seed_sequence])
    input_sequence = input_sequence.reshape(1, len(seed_sequence), 4)

    for _ in range(num_steps):
        prediction = model.predict(input_sequence, verbose=0)
        next_notes = []
        for i in range(4):
            probabilities = prediction[0, i, :]
            predicted_index = np.random.choice(range(num_notes), p=probabilities)
            next_notes.append(index_to_note.get(predicted_index, 0))

        generated_sequence.append(next_notes)
        input_sequence = np.concatenate([input_sequence[:, 1:, :], np.array([next_notes]).reshape(1, 1, 4)], axis=1)

    return generated_sequence

if train_sequences and len(train_sequences[0]) > seq_length:
    start_index = np.random.randint(0, len(train_sequences))
    seed_sequence = train_sequences[start_index][:seq_length]
    num_generate = 100
    generated_music = generate_music(model, seed_sequence, num_generate, note_to_index, index_to_note, num_notes)

    print("\n생성된 음악 (처음 20 스텝):")
    for step in generated_music[:20]:
        print(step)
else:
    print("훈련 데이터가 부족하여 음악 생성을 위한 시드 시퀀스를 얻을 수 없습니다.")
```
